```python
#Rite of words and life
from IPython.display import clear_output
from collections import Counter
import random
class Hangman():
    
    #Variable initialization
    def __init__(self):
        self.w_l = ['beheading', 'hanging', 'boiling', 'quartering', 'the wheel']
        self.e_m = (random.choice(self.w_l))
        self.l_g = ''
        self.g = []
        self.c = 0
        
         
        
    #Lengthy intro, requires 2 positional arguments to run for reason unknown.    
    def printIntro(self, e_m):
        print("""
              In 17th century europe, criminals sentenced to capital punishment were granted\n
              the Rite of Words and Life which granted them a slim chance of dodging their end.\n
              Once strung up to the gallows an executioner would place a five legged stool beneath\n
              their feet, and proceed to draw out five dashes on a board in charcoal.\n 
              As you can imagine, criminals were given five chances to guess the five letters\n
              of the secret phrase. With each wrong guess the executioner would smash a leg out from\n
              the stool with a sledgehammer. Prior to the implementation of a trial by jury,\n
              this was certainly more humane than issuing definite capital punsihment right?\n 
              \t\tUnfortunately the overwhelimg majority of people were illiterate. \n """
        )
    #Game start condition, user can input yae to start, else the game should end
    
    def gameStart(self, e_m):
        ans = input('Would you like your Rite of Words and Life, yae or nae? ')
        if ans == 'yae':
            print('Today\'s topic: Victorian Execution Methods')
            print('_ '*(len(self.e_m)))
            print(self.e_m)
        else:
            print('Thanks for hanging out!')
        
        
    ###Core game logic
    # while the counter is less than the length of the random word +5 the user should be presented with a command to
    #guess a letter, once the user guesses all letters correctly, the program outputs "thanks for hanging out"
    def theGallows(self, c, e_m, l_g, g):

        while self.c < (len(self.e_m) + 5):

            if self.l_g != 'jump':
                self.l_g = input('Guess a letter or type jump to secure your fate, you\'ll get the\n length of the word and 5 extra attempts to guess all letters correctly.')
                self.g.append(self.l_g)


            #quit condition
            elif self.l_g == 'jump':
                print('Thanks for hanging out!')
            
            #loops through each letter in the random word, 
            #if the letter is guessed correctly, it's printed out replacing the _
            for char in self.e_m:
                if char in self.g:
                    print(char, end='')
                
                #prints out a _ for each letter in the random word
                else:
                    print('_', end=' ')
            #increments the counter, and prints an end statement when the counter is equal to
            #the length of the random word
            else:
                self.c = self.c + 1
                print(' ', self.c)
                if self.c == (len(self.e_m)):
                    print(' Thanks for hanging out!')
                  

    def main(self):
        self.__init__()
        
        self.printIntro(self)
        
        self.gameStart(self)
        
        self.theGallows( self.c, self.e_m, self.l_g, self.g)
        

        

###TODO:
#create timer class (import time)
#game difficulty -- implemet timer

game = Hangman()

game.main()
```
